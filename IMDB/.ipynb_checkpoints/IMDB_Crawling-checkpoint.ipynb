음... a 태그를 가지고 있고 href 속성을 가지는 곳에 우리가 찾는 데이터가 있네요

다만 문제는 a 태그는 워낙 흔하기 때문에 단순히 a 태그를 이용해서 검색하는 것도 무리고<br>
우리가 지금까지 해왔던 a 태그에 속성명=속성값 구조로 접근하기에도 난감하네요

왜냐하면 a 태그는 링크 정보를 연결해주는 역할입니다<br>
즉 oldeuboi는 oldeuboi 페이지로 이동하는 링크가 있을 것이고<br>
다른 영화들도 역시나 각자 페이지로 이동하는 링크가 href에 해당 속성값으로 들어가있겠죠

무슨 말이냐면 속성명은 전부 href로 동일하지만 속성값들은 각자의 링크를 지정하였기 때문에 다 값이 다르다는 겁니다<br>
따라서 공통적인 속성값이 없기때문에 <b>태그[속성명=속성값]</b> 구조로 접근할 수 없습니다

---

뭐 말은 어렵게 했지만 간단하게 해결합시다

상위 틀을 잘 살펴보면 h3 태그에 class 이름이 lister-item-header라는 항목이 보일겁니다<br>
이 친구는 다른 곳과 동일하게 되어있는 친구네요!

그렇다면 우리가 잘 알고있는 <b>태그[속성명=속성값]</b>구조로 접근이 가능합니다



```python
meta = soup.select("h3[class=lister-item-header]")
meta
```

그렇다면 h3 태그를 가지고 class 이름이 lister-item-header인 친구들을 찾아봅시다<br>
잘 찾아와지는 지 확인해볼까요?

참고로 class 이름보다는 class 값이 ~다라는게 더 정확한 표현이겠지만<br>
여러분의 쉬운 이해를 위해 단어를 조금 변경한 점 양해부탁드립니다

```python
for i in meta:
    print(i.text)
```

```python
for i in meta:
    print(i.text.strip())
```

이제 meta 안에 우리가 찾고자하는 정보들이 리스트 형태로 받아와졌으니<br>
for문을 통해서 접근합시다

우리가 찾고 싶은 순위 정보와 제목 그리고 덩달아서 연도까지 잘 뽑아진 것을 볼 수 있습니다

---

생각보다 공백이 좀 넓은 게 거슬린다면<br>
strip() 함수를 이용해서 공백을 삭제하셔도 됩니다

---

정말 쉽죠? 지금까지의 코드를 정리해보겠습니다

```python
import requests as r
from bs4 import BeautifulSoup

response = r.get("https://www.imdb.com/list/ls052229572/")

html = response.text
soup = BeautifulSoup(html,"html.parser")

meta = soup.select("h3[class=lister-item-header]")

for i in meta:
    print(i.text)
```

불과 10줄도 안되는 코드를 통해서 데이터를 빠르게 뽑아오는 데 성공하였습니다!

여기서 제가 조금 추가적으로 설명을 드리고자 합니다<br>
만약 여기서 나는 영화 제목만 뽑고싶다고 했을 경우에는 어떻게 해야할까요?

바로 find() 함수를 사용해서 접근합니다

---

다만 find() 함수를 사용하실 때 주의하셔야 할 것은<br>
우리가 아까 받아온 meta 안에서 전체적으로 검색하시는 게 아니라<br>
for문을 이용해 순차적으로 접근한 이후에 검색을 하셔야 한다는 것입니다

말보다는 눈이겠죠! 바로 아래 코드에서 설명하겠습니다

```python
for i in meta:
    print(i.find('a').text)
```

바로 이런식으로 말이죠

다행히도 h3 태그의 class 이름이 lister-item-header라는 항목 안에는<br>
a 태그를 가진 정보가 하나밖에 없어서 매우 잘 실행되는 것을 알 수 있습니다

만약 a 태그를 가진 정보가 여러개라면? 가장 먼저 찾은 값만 가져옵니다<br>
만약 여러개의 정보를 전부 찾고싶다면 find_all() 함수를 쓰시면 됩니다

이처럼 정말 간단한 데이터를 빠르게 긁어와야 하고<br>
웹 크롤링의 기초를 체험하고 싶다면 requests 모듈을 통해 체험하는 것도 좋은 방법입니다
